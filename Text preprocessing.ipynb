```python
import numpy as np
import pandas as pd
from gensim.models import Word2Vec
from gensim.models.word2vec import LineSentence
from gensim.models.word2vec import Word2Vec
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
from sklearn.cluster import KMeans
import re
import random
```

The following two files need to modify the path

```python
#Stopword
with open(r'C:\Users\93531\Desktop\陈彦君\百度停用词表.txt',encoding='utf-8')as f:
    content = f.readlines()
stopwords = [i.replace('\n','') for i in content]
stopwords.append('\\r\\n')
```

```python
df = pd.read_csv(r'C:\Users\93531\Desktop\陈彦君\COMMENTS.csv',encoding='utf-8-sig',engine='python')
```

```python
#Regular expression: if a Chinese character appears in a word, any number or English is retained
#正则表达式的意思是：如果词语中出现了汉字、数字或英文中的任意一种就保留

pattern = re.compile('.*[\w\u4e00-\u9fa5]+.*')
passage =[]
for i in range(df.shape[0]):
    row = df['comments'][i].replace('[\'','').replace('\']','').split('\', \'')
    sentence = []
    for wd in row:
        if re.match(pattern,wd) and wd not in stopwords:
            sentence.append(wd)
    passage.append(sentence)
```

```python
#Check de stop and irrelevant results
#检查去停用词和无关词结果
print(passage[:100])
```

```python
#workers: CPU cores, size: dimensions, and window: how many words are found before and after
#workers表示用多少个CPU核心，size表示多少维度，window表示找前后一共多少个词的关系
model=Word2Vec(passage,hs=1,window=7,size=200,workers=12)
```
